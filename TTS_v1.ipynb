```python
# pip install unidic-lite
# pip install unidic
# python -m unidic download
# You can find more models here: https://tts.readthedocs.io/en/latest/pretrained_models.html

import TTS
from TTS.utils.manage import ModelManager
from TTS.api import TTS
# below needed only if you are playing audio on ipynb
from scipy.io import wavfile
from IPython.display import Audio, display
```

```python
model_manager = ModelManager()
model_path, config_path, model_item = model_manager.download_model('tts_models/en/ljspeech/tacotron2-DDC')

# Load a pre-trained TTS model
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", gpu=False)  # Set gpu=False if no GPU is available
```

```python
def Text_to_speech(text, output_path):
    # Generate speech
    tts.tts_to_file(text=text, file_path=output_path, duration_scaling_factor=1)
```

```python
# Input text to synthesize
text = "Artificial intelligence has transformed the way we interact with technology. From virtual assistants like Siri and Alexa to self-driving cars, AI is seamlessly integrating into our daily lives. But have you ever wondered about the voices behind these technologies? These synthetic voices are the result of years of research in speech synthesis and machine learning. Today, we stand on the brink of making these voices indistinguishable from human speech, opening up possibilities we could only dream of a decade ago."

# Output audio file path
output_path = "/Users/lekshmanannatarajan/Library/CloudStorage/GoogleDrive-zuari1993@gmail.com/My Drive/Colab Notebooks/Coqui_Text_To_Speech/data_local/outputs/output_audio.wav"

Text_to_speech(text, output_path)
# Read the WAV file
rate, audio_data = wavfile.read(output_path)
# Play audio using IPython
display(Audio(data=audio_data, rate=rate, autoplay=True))
```
